## Removing duplicates

```python
# removing duplicates and making it permanent
df.drop_duplicates(inplace = True)
```

```python
# Checking the duplicate values
df.duplicated().sum()
```

## Handling missing values

#### Handle DisNo. column

##### To handling DisNo. column we are going to drop the null values.
- Because DisNo. it is the feature where disaster id or number is stored.
- We can't fill these values 
- Disaster numbers are unique values that mentioned the disasters occurs.
- so this means there is no data for these disaster number feature.

```python
# droping null values from Disaster number column permantly
df.dropna(axis=0,subset='DisNo.',inplace=True)
```

```python
# Checking the Null values count
df.isnull().sum()
```

#### Handle Location column

##### To handling location column we have to take following steps:
- we use mode to fill the null values.

```python
# Filling null values by mode because its in the category 
mode_loc = df['Location'].mode()[0]
df['Location'].fillna(mode_loc,inplace = True)
```

```python
#checking the null values
df.isnull().sum()
```

#### Handle Associated Types column

##### To handling Associated Types column we have to take following steps:
- we use mode to fill the null values.

```python
# filling null values in associated types column by mode
mode_astype = df['Associated Types'].mode()[0]
df['Associated Types'].fillna(mode_astype,inplace = True)
```

```python
# checking null values
df.isnull().sum()
```

#### Handle Insured Damage ('000 US$) Types column

##### To handling Insured Damage ('000 US$) column we have to take following steps:
1.Check distribution of column
- If it is normally distributed then impute it with mean value
- If not, It means Data is skewed So, we have to impute with median

```python
# Checking its normally distributed or not for filling null values
plt.figure(figsize=[5,3])
sns.distplot(df["Insured Damage ('000 US$)"])
#plt.show()
```

```python
# using median for filling null values because its not normal distribution
med = df["Insured Damage ('000 US$)"].median()
df["Insured Damage ('000 US$)"].fillna(med,inplace = True)
```

```python
# checking null values
df.isnull().sum()
```

#### Handle Total Damage ('000 US$) column

##### To handling Total Damage ('000 US$) column we have to take following steps:
1.Check distribution of column
- If it is normally distributed then impute it with mean value5
- If not, It means Data is skewed So, we have to impute with median

```python
# Plotting the distribution plot
plt.figure(figsize=[5,3])
sns.distplot(df["Total Damage ('000 US$)"])
plt.show()
```

```python
# filling null values with median
medi = df["Total Damage ('000 US$)"].median()
df["Total Damage ('000 US$)"].fillna(medi,inplace = True)
```

```python
# Checking the null values
df.isnull().sum()
```

#### Handle CPI column

##### To handling CPI column we have to take following steps:
1.Check distribution of column
- If it is normally distributed then impute it with mean value
- If not, It means Data is skewed So, we have to impute with median

```python
# checking Data is skewed or not
plt.figure(figsize = [5,3])
sns.distplot(df["CPI"])
plt.show()
```

```python
# imputing null values with mean
mean = df["CPI"].mean()
df['CPI'].fillna(mean,inplace = True)
```

```python
# checking the null values
df.isnull().sum()
```

```python
# dropping null values in Start year column
df.dropna(axis=0,subset='Start Year',inplace=True)
```

```python
# dropping null values in end year column
df.dropna(axis=0,subset='End Year',inplace=True)
```

```python
# checking null values
df.isnull().sum()
```

### Fixing the data types of year column

```python
# convert data type of start year
df['Start Year'] = pd.to_numeric(df['Start Year'], errors='coerce')
df['Start Year'] = df['Start Year'].astype(int)
```

```python
# convert data type of end year
df['End Year'] = pd.to_numeric(df['End Year'], errors='coerce')
df['End Year'] = df['End Year'].apply(int)
```

## Data Visualization

```python
# return first five rows
df.head()
```

### Q1 Disaster types that happened most

```python
# Applying a color palette 
sns.set_palette('viridis')
```

```python
# plotting a countplot 
plt.figure(figsize = [8,3])                        
sns.countplot(x=df['Disaster Type'])               
plt.xticks(rotation = 45)                          
plt.title("Frequency of Disaster types")
plt.xlabel("Disaster Type")
plt.ylabel("Frequency")
plt.show()
```

#### Insights

- The bar chart reveals that floods are the most frequently occurring natural disaster
- In contrast, events like mass movements (both wet and dry), glacier lake outbursts, extreme temperatures, and wildfires are among the least common disasters.

### Q2. Disasters Subgroup 

```python
#plotting a pie chart
plt.figure(figsize = [5,4])
df['Disaster Subgroup'].value_counts().plot.pie(autopct = '%2.2f%%')
plt.title("Distribution of Disaster subgroups")
plt.show()
```

#### Insights

- The pie chart indicates that hydrological and meteorological subgroups are the most frequently occurring disasters
- while climatological and geophysical subgroups are the least common.

### Q3. Damages by Disaster Subgroup

```python
# Plotting a Scatter plot
plt.figure(figsize = [5,4])
sns.scatterplot(x = df["Insured Damage ('000 US$)"].values,y = df["Total Damage ('000 US$)"].values,hue = df['Disaster Subgroup'])
plt.show()
```

#### Insights

- The scatter plot illustrates that insured damages ranging from 0 to 200,000 are average.
- Hydrological disasters demand the highest insured damages
- while biological and geophysical disasters require the least.

### Q4.Top 5 locations

```python
# Plotting Horizantal bar visual
plt.figure(figsize = [5,3])
data = df['Location'].value_counts().head(10)
data.plot(kind='barh')
plt.title("Top 10 Locations That Suffered")
plt.show()
```

#### Insights

- The horizontal bar chart highlights that Uttar Pradesh and Andhra Pradesh have experienced the highest number of disasters
- while Delhi and Gujarat have faced the least.

### Q5.CPI and Total Deaths trends

```python
# plotting 2 subplots
fig,axes = plt.subplots(1,2,figsize=(12,3))
axes[0].plot(df['Start Year'],df['CPI'])
axes[0].set_xlabel("Start Year")
axes[0].set_ylabel("CPI")
axes[0].set_title("Start Year wise CPI")

axes[1].plot(df['End Year'],df['Total Deaths'])
axes[1].set_xlabel("End Year")
axes[1].set_ylabel("Total Deaths")
axes[1].set_title("End Year wise Total Deaths")
plt.show()
```

#### Insights

- The first subplot shows a noticeable growth in the CPI (Consumer Price Index) after 1960
- In the second subplot, the period between 1960 and 1980 records the highest number of total deaths
- at the end of disasters, with another significant peak observed in 2020.

### Q6. Disaster trends over the years

```python
#plotting lineplot
plt.figure(figsize = [8,3])
sns.lineplot(x = df['Disaster Type'],y = df['Start Year'].values,marker='o')
plt.title("Disaster Type over years")
plt.ylabel("Years")
plt.xlabel("Disaster Types")
plt.xticks(rotation = 45)
plt.show()
```

#### Insights

- The line plot indicates that glacial lake outburst floods were most frequent in 2020
- Between 1980 and 2000, all types of disasters occurred, but after 2010, there has been a noticeable increase in glacial lake outburst floods.

### Q7. Distribution of Dis Mag Value

```python
#creating histogram
plt.figure(figsize = [7,4])
sns.histplot(df["Dis Mag Value"],kde=True,bins=40)
plt.xlabel("Dis Mag Value")
plt.ylabel("Frequency")
plt.title("Distribution of Dis Mag Value")
plt.show()
```

#### Insights

- The histogram shows that the majority of disaster magnitude (dis mag) values fall within the range of 0 to 100,000, indicating a concentration of lower magnitude events in this range.

### Q8. Number of people suffered in Disaster

```python
# creating violin plot
plt.figure(figsize=[10,3])
sns.violinplot(x=df["Disaster Type"],y=df["No Suffered"])
plt.xticks(rotation=45)
plt.show()
```

#### Insights

- The violin plot reveals that disasters such as droughts, earthquakes, and storms caused the most significant suffering among people
- In contrast, events like wildfires, mass movements, and glacial lake outbursts impacted the least number of people.

### Q9.Number of homeless people by Disaster subgroup

```python
# creating jointplot
plt.figure(figsize=[5,3])
sns.jointplot(x=df["Disaster Subgroup"],y=df["No Homeless"])
plt.xticks(rotation=45)
plt.show()
```

#### Insights

- The joint plot indicates that hydrological and meteorological disasters are associated with a higher number of people left homeless compared to other disaster subgroups.

### Q10. Disaster type vs CPI

```python
# plotting boxenplot
plt.figure(figsize=[8,4])
sns.boxenplot(x=df["Disaster Type"],y=df["CPI"])
plt.xticks(rotation=45)
plt.show()
```

#### Insights

- Disasters like Floods and Storms show a high range of CPI values, indicating that their economic impacts vary significantly.
- Disasters like Wildfires, Infestations, and Glacial Lake Outburst Floods generally show lower CPI values
- Floods and storms tend to have the most variable and potentially severe economic impacts (CPI)

### Q11. Total damage by disaster

```python
# Plotting pointplot
plt.figure(figsize=[8,3])
sns.pointplot(x=df["Disaster Subgroup"],y=df["Total Damage ('000 US$)"])
```

#### Insights

- The chart indicates that climatological and meteorological events result in significantly higher total damage 
- compared to geophysical and hydrological events, which have a more moderate impact.
